```python
# Crear una onda gravitacional
#   mediante un arreglo numpy

import numpy as np
import matplotlib.pyplot as plt

from scipy import signal as scisignal
#from constantQ.timeseries import TimeSeries
from gwpy.timeseries import TimeSeries

# Generar un np.array de la GW
dt = 0.001
t = np.arange(0,3,dt)
f0 = 50
f1 = 250
t1 = 2
x = np.cos(2*np.pi*t*(f0 + (f1 - f0)*np.power(t, 2)/(3*t1**2)))
fs = 1/dt

plt.plot(x)				# graficar la señal
plt.show()				# mostrar
```

```python
series = TimeSeries(x, dt = 0.001, unit='m', name='test', t0=0)     #np.array --> constantQ.timeseries   
sq = series.q_transform(search=None)				# q transform
print(len(sq[0]))       # freq array length
print(len(sq))          # time array length


plt.imshow(sq.T, origin='lower')				# graficar el espectrograma
plt.colorbar()									
plt.show()										# mostrar
```

```python
data = TimeSeries.fetch_open_data('H1', 1126259446, 1126259478)

qspecgram = data.q_transform(outseg=(1126259462.2, 1126259462.5))
plot = qspecgram.plot(figsize=[8, 4])
ax = plot.gca()
ax.set_xscale('seconds')
ax.set_yscale('log')
ax.set_ylim(20, 500)
ax.set_ylabel('Frequency [Hz]')
ax.grid(True, axis='y', which='both')
ax.colorbar(cmap='viridis', label='Normalized energy')
plot.show()
```

```python
freq, ts, Sxx = scisignal.spectrogram(qspecgram)		# scipy spectrogram

plt.pcolor(ts, freq, Sxx)
plt.colorbar()
plt.show()
```
